```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#distribution of Legit and fraud transactions
df['Class'].value_counts()
```

About the data

This data is highly unbalanced


0 ---> normal Transactions

1 ---> Fraud transaction

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
fraud.info()
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values of both transactions
df.groupby('Class').mean()
```

Due to Unbalanced data we have to sample some data from legit dataset so the ML Model can make unbiased predictions


Total number of fraud data --> 492

```python
legit_sample=legit.sample(n=492)
```

```python
legit_sample
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

Now the Data is uniform

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(['Class'], axis='columns')
y=new_dataset['Class']
```

```python
y
```

Split the data into train and split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, stratify=y, random_state=2)
```

Model Training Using Logistic Regression

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

Model Evalution

```python
#accuracy on training data
x_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Model Accuracy Score : ",training_data_accuracy)
```

```python
#Accuracy of Test Data

x_test_prediction=model.predict(X_test)
```

```python
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Model Accuracy Score: ",test_data_accuracy)
```
